```python
import random
import numpy as np
np.random.seed(10) #fixed seeds
m=np.random.randint(1, high=4, size=(5, 3)) #generate 5*3 matrix
#m = np.random.rand(5,3)
print(m)
n=0
while n<10:
  #crossover
  for i in range(0,4):
     
    m1=m[i]
    m2=m[i+1]
    print(m1)
    
    print(m2)
    crossover1(m1,m2)
  
  #mutation 
  A=0
  B=1#小数的范围A ~ B
  a=random.uniform(A,B)
  C=1#随机数的精度round(数值，精度)
  mutation_prob = round(a,C)
  #p is chosen to be the mutation point index
  p =np.random.randint(0, high=4)
  if mutation_prob == 0.1:
      mutation(m[p])
  # delete the worst 2 gene
  list = []
  for j in range(5):
    list.append(fun(m[j]))  #get all the final value of m[0],m[1],..m[5]
  idx_worst = []
  for k in range(5):
    if fun(m[k]) == max(list):
      idx_worst.append(k)
  for l in range(len(idx_worst)):
    m[idx_worst[l]] = np.random.randint(1, high=4, size=(1, 3))
  n+=1
ans = []
for i in range(5):
  ans.append(fun(m[i]))
#  print(fun(m[i]))
print(min(ans))



```
